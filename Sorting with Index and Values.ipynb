```python
df['Height_clean'] = df['Height'].replace(['NaN', '...'], np.nan)

# 3. Split the 'feet-inches' string into two separate columns
# 'expand=True' creates new columns ('0' for feet, '1' for inches)
parts = df['Height_clean'].str.split('-', expand=True)

# 4. Convert the new 'feet' and 'inches' columns to numeric types
# 'errors='coerce'' will turn any errors (like on a NaN row) into NaN
feet = pd.to_numeric(parts[0], errors='coerce')
inches = pd.to_numeric(parts[1], errors='coerce')

# 5. Apply the conversion formula: (Feet * 12 + Inches) * 2.54
total_inches = (feet * 12) + inches
df['Height_cm'] = total_inches * 2.54
```
