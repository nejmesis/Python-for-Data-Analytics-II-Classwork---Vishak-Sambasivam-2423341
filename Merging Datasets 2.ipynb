```python
import pandas as pd

def load_csv(file_path):
    return pd.read_csv(file_path)

food = r"C:\Users\User\Downloads\SBA-class-activity-main\restaurant_foods.csv"
rw1s = r"C:\Users\User\Downloads\SBA-class-activity-main\restaurant_week_1_sales.csv"
rw2s = r"C:\Users\User\Downloads\SBA-class-activity-main\restaurant_week_2_sales.csv"
rw1t = r"C:\Users\User\Downloads\SBA-class-activity-main\restaurant_week_1_times.csv"
rcust = r"C:\Users\User\Downloads\SBA-class-activity-main\restaurant_customers.csv"

food_data = load_csv(food)
print('FOOD')
display(food_data.head(5))

res_w1_sales = load_csv(rw1s)
print('WEEK 1 SALES')
display(res_w1_sales.head(5))

res_w2_sales = load_csv(rw2s)
print('WEEK 2 SALES')
display(res_w2_sales.head(5))

customer_data = load_csv(rcust)
print('CUSTOMERS')
display(customer_data.head(5))

res_w1_time = load_csv(rw1t)
print('WEEK 1 TIME')
display(res_w1_time.head(5))
```

```python
res_w2_sales.merge(customer_data, how='left', left_on='Customer ID', right_on='ID')
```

```python
res_w2_sales.merge(customer_data, how='left', left_on='Customer ID', right_on='ID').drop('ID', axis='columns')
```

```python
res_w2_sales.merge(customer_data, how='inner', left_on='Customer ID', right_on='ID').drop('ID', axis='columns')
```

### Inner Join

It combines rows from both dataframes only when there are matching values in a specified common column (or index).

```python
res_w1_sales[res_w1_sales['Customer ID'] == 155]
```

```python
res_w2_sales[res_w2_sales['Customer ID'] == 155]
```

### Finding the changes in sales with merge function

```python
res_w1_sales.merge(res_w2_sales, how='inner', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'])
```

### Merging two dataframes with common columns

```python
res_w1_sales.merge(res_w2_sales, how='inner', on=['Customer ID', 'Food ID'])
```
