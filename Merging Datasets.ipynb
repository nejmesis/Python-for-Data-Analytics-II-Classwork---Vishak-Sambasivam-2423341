### Concat function

The `pd.concat()` function in the pandas library is used to concatenate or combine multiple Series or DataFrame objects along a specific axis (either rows or columns). 

This is particularly useful for stacking datasets on top of each other or stitching them side-by-side to create a unified dataset for analysis.  

`pd.concat([df1, df2])`

`pd.concat([df1, df2], ignore_index = False)`

`pd.concat([df1, df2], ignore_index = True)`

`pd.concat([df1, df2], keys = ["df1", "df2"]).sort_index()`

```python
pd.concat([res_w1_sales, res_w2_sales])
```

```python
print('Concat with ignore_index = False')
display(pd.concat([res_w1_sales, res_w2_sales], ignore_index = False))

print('Concat with ignore_index = True')
display(pd.concat([res_w1_sales, res_w2_sales], ignore_index = True))
```

```python
print('Concat with sort_index()')
display(pd.concat([res_w1_sales, res_w2_sales], keys = ["res_w1_sales", "res_w2_sales"]).sort_index())
```

```python
df1 = pd.DataFrame([1,2,3], columns=['A'])
df2 = pd.DataFrame([4,5,6], columns=['B'])
```

```python
print('Merging the columns by stacking the dataframes')
display(pd.concat([df1, df2]))
```

```python
print('Merging the columns side by side')
display(pd.concat([df1, df2], axis='columns'))
```

### Merge function - Left Join 

`df1.merge(df2, how='left', on='common attribute')`

```python
print('Merging FOOD Data to WEEK 1 SALES Data')
display(res_w1_sales.merge(food_data, how='left', on="Food ID"))
```

### Merge function - Right Join

`df1.merge(df2, how='right', on='common attribute')`

```python
res_w2_sales.merge(food_data, how='right', on='Food ID')
```

### More on left join

`df1.merge(df2, how='left', left_on='common attribute', right_on='other common attribute')`

`df1.merge(df2, how='left', left_on='common attribute', right_on='other common attribute').drop("other common attribute", axis='')`

```python
res_w2_sales.merge(customer_data, how='left', left_on='Customer ID', right_on='ID')
```

```python
res_w2_sales.merge(customer_data, how='left', left_on='Customer ID', right_on='ID').drop('ID', axis='columns')
```
